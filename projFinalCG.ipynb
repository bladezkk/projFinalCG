```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

thermal1 = np.array(Image.open('thermal1.png'))[:, :, :3]
thermal2 = np.array(Image.open('thermal2.png'))[:, :, :3]
mvt = thermal2.astype(np.int16) - thermal1.astype(np.int16)
mvt = np.clip(mvt, 0, 255).astype(np.uint8)

plt.figure(figsize=(16, 16))
plt.subplot(2, 2, 1)
plt.imshow(thermal1)
plt.subplot(2, 2, 2)
plt.imshow(thermal2)
plt.subplot(2, 2, 3)
plt.imshow(mvt)
```

```python
urso1 = np.array(Image.open('urso1.png'))[:, :, :3]
urso2 = np.array(Image.open('urso2.png'))[:, :, :3]
mvt = urso1.astype(np.int16) - urso2.astype(np.int16)
mvt = np.clip(mvt, 0, 255).astype(np.uint8)

plt.figure(figsize=(16, 16))
plt.subplot(2, 2, 1)
plt.imshow(urso1)
plt.subplot(2, 2, 2)
plt.imshow(urso2)
plt.subplot(2, 2, 3)
plt.imshow(mvt)
```
